<b>Problem Statement :</b><br>In this assignment students have to transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
np.random.seed(5)
iris = datasets.load_iris()
x = iris['data']
y = iris['target']
```

```python
fig = plt.figure(1,figsize = (6,5))
plt.clf()

ax = Axes3D(fig, rect = [0,0,0.95,1], elev = 50, azim = 140)

plt.cla()
pca = decomposition.PCA(n_components = 3)
pca.fit(x)
x = pca.transform(x)

for name,label in [('Setosa',0),('Versicolour',1),('Virginica',2)]:
    ax.text3D(x[y==label,0].mean(),
              x[y==label,1].mean() + 1.5,
              x[y==label,2].mean(), name, horizontalalignment = 'center', 
              bbox = dict(alpha=.5, edgecolor='w', facecolor = 'w'))
    
y = np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(x[:,0],x[:,1],x[:,2],c=y, cmap='hsv', edgecolor = 'k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()

```
